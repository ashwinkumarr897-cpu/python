```python
# Ladder if...else - Multiple Conditions Checked one by one

a=int(input("Enter the A Value : "))
b=int(input("Enter the B Value : "))
c=int(input("Enter the C Value : "))

if(a>b and a>c):
  print("A is Biggest Value")
elif(b>a and b>c):
  print("B is Biggest Value")
elif(c>a and c>b):
  print("C is Biggest Value")
else:
  print("A,B,C is Same Value")
```

```python
# Nested if...else - An if inside another if

a=int(input("Enter the A Value : "))
b=int(input("Enter the B Value : "))
c=int(input("Enter the C Value : "))

if(a>b):
  if(a>c):
    print("A is Biggest Value")
  else:
    print("C is Biggest Value")
else:
  if(b>c):
    print("B is Biggest Value")
  else:
    print("C is Biggest Value")
```

```python
# match Statement - Python 3.10+ - Works like Switch Case

char=input("Enter the Character - Vowel - a,e,i,o,u : ")

match char:
  case 'A':
    print("Vowel")
  case 'E':
    print("Vowel")
  case 'O':
    print("Vowel")
  case 'I':
    print("Vowel")
  case 'U':
    print("Vowel")
  case 'a':
    print("Vowel")
  case 'e':
    print("Vowel")
  case 'o':
    print("Vowel")
  case 'i':
    print("Vowel")
  case 'u':
    print("Vowel")
  case _:
    print("Not Vowel")
```

```python
num1=int(input("Enter the Number 1 : "))
num2=int(input("Enter the Number 2 : "))

opr=input("Enter the Operator : (+,-,*,/,**) : ")

match opr:
  case '+':
    answer=num1+num2
    print(answer)
  case '-':
    answer=num1-num2
    print(answer)
  case '*':
    answer=num1*num2
    print(answer)
  case '/':
    answer=num1/num2
    print(answer)
  case '**':
    answer=num1**num2
    print(answer)
  case _:
    print("Invalid Input Please Try Again")
```

```python
# strip() - Remove Spaces
# lstrip() - Remove Left Spaces
# rstrip() - Remove Right Spaces

S="       Sridhar      "
print(S)

print(S.strip())
```

```python
# repace() - Replace SubString

text="Call Cat"
print(text.replace("C", "B"))
```

```python
# find() - Find Position

text="Python is Powerful"
print(text.find("ful"))
```

```python
# count() - Count Occurrences

text="Sridhar Raj P"
print(text.count("a"))
```

```python
text="Python is Powerfull"
print(text.startswith("Py"))
print(text.endswith("full"))
```

```python
# split() - Convert to List
Names="Sridhar Abi Priya Kumar Ravi Rani"
print(Names.split(" "))
```

```python
# join() - Join list to String
topics=["Core", "OOPS", "DS", "Database"]
print("-".join(topics))
```

```python
# isalpha() - only letter
check="Sridhar"
print(check.isalpha())
```

```python
# isdigit() - Only Digits
check="1997"
print(check.isdigit())
```

```python
# isalnum() - Letters + Digits

check="Sridhar123"
print(check.isalnum())
```

```python
check="sridhar"
print(check.islower())
```

```python
check="SRI"
print(check.isupper())
```

```python
# Reverse a String

S=input("Enter a String : ")
print(S[::-1])
```

```python
# Check Palindrome

S=input("Enter a String : ")
print("Palindrome" if S==S[::-1] else "Not Palindrome")
```

```python
# Count Vowels

S=input("Enter a String : ")
count=0
for i in S:
  if i.lower() in 'aeiou':
    count+=1
print(count)
```
